## Importing all the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255
```

## Model Building - (Arch 1)

```python
model = Sequential(
    name = 'model_arch_1',
    layers = [Flatten(input_shape = (28,28)),
              Dense(128, activation = 'relu'),
              Dense(10,activation = 'softmax')]
)
```

```python
model.summary()
```

```python
model.compile(loss = 'sparse_categorical_crossentropy',
              optimizer = 'adam',
              metrics = 'accuracy')
```

```python
history = model.fit(
    x=X_train,
    y=y_train,
    batch_size=None,
    epochs=10,
    verbose=1,
    callbacks=None,
    validation_split=0.20,
    validation_data=None,
    shuffle=True,
    class_weight=None,
    sample_weight=None,
    initial_epoch=0,
    steps_per_epoch=None,
    validation_steps=None,
    validation_batch_size=None,
    validation_freq=1,
    max_queue_size=10,
    workers=1,
    use_multiprocessing=False,
)
```

```python
history.history
```

```python
def plot_accuracy_loss():

    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.title('Accuracy')
    plt.plot(np.arange(0,len(history.history['accuracy'])),history.history['accuracy'], label = 'Training')

    plt.plot(np.arange(0,len(history.history['val_accuracy'])),history.history['val_accuracy'], label = 'Validation')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10,4))
    plt.subplot(1,2,2)
    plt.title('Loss')
    plt.plot(np.arange(0,len(history.history['loss'])),history.history['loss'], label = 'Training')

    plt.plot(np.arange(0,len(history.history['val_loss'])),history.history['val_loss'], label = 'Validation')
    plt.legend()
    plt.show()

    



plot_accuracy_loss()
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis = 1)
```
